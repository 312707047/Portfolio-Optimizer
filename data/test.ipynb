```python
import crypto
import os
from multiprocessing import Pool
import itertools
```

```python
def get_data(tag):
    crypto.get_all_binance(f'{tag}USDT', '15m')
```

```python
pool = Pool(os.cpu_count())

pool.map(get_data, tag)
```

```python
tag = ['XRPUSDT', 'SOLUSDT', 'DOTUSDT', 'LTCUSDT', 'MATICUSDT', 'LINKUSDT', 'BCHUSD']
tag = list(itertools.zip_longest(tag, ['15m'], fillvalue='15m'))
print(tag)
with Pool() as pool:
    pool.starmap(crypto.get_all_binance, tag)
```
